```python
import random
attempts = 3
correct_num = random.randint(1, 101)


while attempts >= 1 and attempts <= 3:
    guess_attempt = int(input("Guess a number ü§∑üèø‚Äç‚ôÇÔ∏è: "));

    if guess_attempt < correct_num:
        print(f"{guess_attempt} is low ‚ùå .")
    elif guess_attempt > correct_num:
        print(f"{guess_attempt} is high ‚ùå.")
    else:
        print("You are correct")
        break

    print("Try Again! üò©")
    attempts -= 1

print(f"The correct answer is {correct_num} ‚úÖ.")
```

```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
operator_list = ["+", "-", "*", "/"]
operand = input("Enter the operator (+, -, *, /): ")

while operand not in operator_list:
    operand = input("Enter the operator (+, -, *, /): ")

answer  = 0
if operand == "+":
    answer = num1 + num2
elif operand == "-":
    answer = num1 - num2
elif operand == "*":
    answer = num1 * num2
elif operand == "/":
    answer = num1 / num2

print(f"Answer: {answer}")
```

```python
def print_info(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

print_info(name="Jane", age=25)
```

```python
def square(x):
    return x * x

def display_square(n):
    result = square(n)
    print("square:", result)

display_square(20)
```

```python
def add(*numbers):
    total = 0
    for number in numbers:
        total += number
    
    return total

def difference (a, b):
    return a - b


def mul (a, b):
    return a * b

def div (a, b):
    if b == 0:
        return "cannot divide"
    else:
        return a / b

a = float(input("Num1:"))
b = float(input("Num2:"))
operation = input("Enter operation (+, -, /, *): ")

if operation == "+":
    print(f"result: {add(a, b)}")
elif operation == "-":
    print(f"result: {difference(a, b)}")
elif operation == "*":
    print(f"result: {multiply(a, b)}")
elif operation == "/":
    print(f"result: {add(a, b)}")
else:
    print("operation not Identified")
```
