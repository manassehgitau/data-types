### sample

```python
import csv 
import requests
from bs4 import BeautifulSoup

base_url = 'https://quotes.toscrape.com/page/{}/'

csv_file= "quotes.csv"
with open(csv_file, mode='w', newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Quote", "Author"])

    page = 1
    while True:
        url = base_url.format(page)
        response = requests.get(url)

        if response.status_code != 200:
            print(f"Error to retrieve page {page}. status code: {response.status_code}")
            break

        soup = BeautifulSoup(response.text, 'html.parser')
        quotes= soup.select(".quote")

        if not quotes:
            break

        for quote in quotes:
            text = quote.select_one(".text").get_text()
            author = quote.select_one(".author").get_text()
            writer.writerow([text, author])
        page += 1

print(f"All Quotes saved to {csv_file}")
```

### Quotes

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

# Set up ChromeDriver service
service = Service()  # If chromedriver is in PATH
driver = webdriver.Chrome(service=service)

# Open a website
driver.get("https://quotes.toscrape.com")
quotes =driver.find_elements(By.CLASS_NAME, "quote")

for quote in quotes:
    text= quote.find_element(By.CLASS_NAME, "text").text
    author = quote.find_element(By.CLASS_NAME, "author").text
    print(f"{text} by {author}\n")
    print("=================================\n")

time.sleep(5)

driver.quit()
```

### Books

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

# Set up ChromeDriver service
service = Service()  # If chromedriver is in PATH
driver = webdriver.Chrome(service=service)

# Open a website
driver.get("https://books.toscrape.com/")
books =driver.find_elements(By.CLASS_NAME, "product_pod")

# print(books)

for book in books:
    book_name= book.find_element(By.TAG_NAME, "h3").find_element(By.TAG_NAME, "a").get_attribute("title")
    book_price = book.find_element(By.CLASS_NAME, "price_color").text
    book_in_stock = book.find_element(By.CLASS_NAME, "instock").text
    print(f"{book_name}\n")
    print(f"{book_price}\n")
    print(f"{book_in_stock}\n")
    print("=================================\n")

time.sleep(5)

driver.quit()
```

### scrape for all  pages and save to csv

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import csv

books_base_url = "https://books.toscrape.com/catalogue/page-{}.html"
csv_file = "selenium_books.csv"

# Set up ChromeDriver service
service = Service()  # If chromedriver is in PATH
driver = webdriver.Chrome(service=service)

with open(csv_file, mode='w', newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "Price", "inStock"])


    page = 1 
    while True:
        # Open a website
        url = books_base_url.format(page)
        driver.get(url)
        books =driver.find_elements(By.CLASS_NAME, "product_pod")
        # print(books)

        if not books:
            break


        for book in books:
            book_name= book.find_element(By.TAG_NAME, "h3").find_element(By.TAG_NAME, "a").get_attribute("title")
            book_price = book.find_element(By.CLASS_NAME, "price_color").text
            book_in_stock = book.find_element(By.CLASS_NAME, "instock").text
            writer.writerow([book_name, book_price, book_in_stock])


        page += 1

    time.sleep(5)

    driver.quit()
```
